```python
X = df.drop('SALARY',axis=1)
y = df['SALARY']
```

```python
X
```

```python
# Model Building
trf = ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(drop='first',sparse=False),[0,1,2]),
    ('std_scalar',PowerTransformer(),[4,5,6,7,9,10]),
    ('Min_max',MinMaxScaler(),[8])
],remainder='passthrough')
```

```python
def model_score(model):
    pipe = Pipeline(steps=[
        ('transform',trf),
        ('model',TransformedTargetRegressor(regressor=model()))
    ])

    print('Cross_validation_score',-cross_val_score(pipe,X,y,cv=5,n_jobs=2,verbose=True,scoring='neg_root_mean_squared_error').mean())
    print('Cross_validation_score',cross_val_score(pipe,X,y,cv=5,n_jobs=2,verbose=True,scoring='r2').mean())
    print('Cross_validation_score',-cross_val_score(pipe,X,y,cv=5,n_jobs=2,verbose=True,scoring='neg_mean_squared_error').mean())
```

```python
model_score(RandomForestRegressor)
```

```python
# Neural network
model = Sequential()

model.add(Dense(128,activation='relu',input_dim=34))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='linear'))

model.compile(loss='mae',optimizer='adam',metrics=[RootMeanSquaredError()])
```

```python
model.summary()
```

```python
history = model.fit(x=trf.fit_transform(X),y=y,batch_size=20,epochs=250,validation_split=.3)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['root_mean_squared_error'])
plt.plot(history.history['val_root_mean_squared_error'])
```
